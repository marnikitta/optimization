```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import subprocess
from sklearn.datasets import load_svmlight_file
```

```python
filename = 'gisette_scale'
```

```python
train_X, train_target = load_svmlight_file('datasets/' + filename)
```

```python
train_X.shape
```

```python
train = pd.DataFrame(train_X.toarray())
train['target'] = train_target
train.head()
```

```python
train.to_csv(filename + '.tsv', sep='\t', header=False)
```
